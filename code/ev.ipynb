#### <p style="text-align: center;"> Function to get vector data </p>

```python
def dist(x1, y1, x2, y2):
    return np.sqrt((x1-x2)**2 + (y1-y2)**2)

def get_attr_area(table, attr, lat, lng, distance=0.5, first=False):
    new_lat  = [lat  - (distance / 6378.137) * (180 / math.pi), lat  + (distance / 6378.137) * (180 / math.pi)];
    new_lng = [lng - (distance / 6378.137) * (180 / math.pi) / math.cos(lng * math.pi/180), lng + (distance / 6378.137) * (180 / math.pi) / math.cos(lng * math.pi/180)]
    result = table.query(f'{new_lat[0]} <= lat <= {new_lat[1]} & {new_lng[0]} <= lng <= {new_lng[1]} ').copy()
    if first:
        result['dist'] = result.apply(lambda x: dist(x.lng, x.lat, lng, lat), axis=1)
        result.sort_values('dist', inplace=True)
        return result.iloc[0][attr]
    return result[attr]

# add nearby population count for each charger
allData['pop_near'] = allData.apply(lambda x: get_attr_area(population, 'pop', x.lat, x.lng, 0.5).sum(), axis=1)
allData
```

```python
# add nearby parking space count
allData['park_near'] = allData.apply(lambda x: get_attr_area(parking, 'lat', x.lat, x.lng, 0.5).size, axis=1)
allData
```

```python
# add land-use classification
allData['class'] = allData.apply(lambda x: get_attr_area(umweltAtlas, 'usage', x.lat, x.lng, 0.5, first=True), axis=1)
allData['class_desc'] = allData.apply(lambda x: get_attr_area(umweltAtlas, 'usage_plain', x.lat, x.lng, 0.5, first=True), axis=1)
allData
```

###### Since the data is not evenly distributed, there is not a one-size-fits-all search range. This implementation tries again when there are no results with the `first` flag.

```python
def get_attr_area2(table, attr, lat, lng, first=False, distance=0.5):
    new_lat  = [lat  - (distance / 6378.137) * (180 / math.pi), lat  + (distance / 6378.137) * (180 / math.pi)];
    new_lng = [lng - (distance / 6378.137) * (180 / math.pi) / math.cos(lng * math.pi/180), lng + (distance / 6378.137) * (180 / math.pi) / math.cos(lng * math.pi/180)]
    result = table.query(f'{new_lat[0]} <= lat <= {new_lat[1]} & {new_lng[0]} <= lng <= {new_lng[1]} ').copy()
    if first:
        result['dist'] = result.apply(lambda x: dist(x.lng, x.lat, lng, lat), axis=1)
        result.sort_values('dist', inplace=True)
        if len(result)>= 1:
            return result.iloc[0][attr]
        else:
            return get_attr_area2(table, attr, lat, lng, first, distance=50)
    return result[attr]
```

```python
def OTH(allData,col_name, low_quantile=0.25,up_quantile=0.75):
    Q1=allData[col_name].quantile(low_quantile)
    Q3=allData[col_name].quantile(up_quantile)
    IQR=Q3-Q1
    low_limit=Q1-1.5*IQR
    up_limit=Q3+1.5*IQR
    print(col_name,low_limit,up_limit)
    return low_limit,up_limit
def replace(dataframe,variables):
    low_limit, up_limit= OTH(dataframe,variables)
    dataframe.loc[(dataframe[variables]<low_limit),variables]=low_limit
    dataframe.loc[(dataframe[variables]>up_limit),variables]=up_limit
```

```python
def detection(dataframe,col_name):
    low_limit,up_limit=OTH(dataframe,col_name)
    if dataframe[(dataframe[col_name]>up_limit)| (dataframe[col_name]<low_limit)].any(axis=None):
        return True
    else:
        return False
```

##### AHP + Unsupervised learning
The eigenvectors were calculated manually

```python
def calc_score(x):
    score = 0
    weights = [0.3419, 0.2279, 0.1899, -0.0949, 0.0791]
    i = 0
    for col in x:
        if i == 30:
            break
        elif i%6 == 2:
            score += col * weights[i//6]
        i += 1

    return score
```

```python
detailedClusters = allData_copy[['adoption','traffic_near','pop_near','charger_near', 'park_near', 'class','cluster']]\
.groupby('cluster')\
.agg({'count','std','max', 'mean','min', 'median'}).copy()

clusters = pd.DataFrame([0, 1, 2, 3, 4, 5, 6], columns=['cluster'])

clusters['score'] = detailedClusters.apply(calc_score, axis=1)
clusters = clusters.sort_values('score')
clusters['score'] = [0.1, 0.15, 0.3, 0.7, 0.3, 0.15, 0.1]
clusters.sort_values('score')
```

```python
c_locs = locs[['lat', 'lng', 'cluster']].copy()
c_locs = pd.merge(c_locs, clusters, how='outer', on='cluster')
c_locs = c_locs.drop(columns=['cluster'])
c_locs
```
